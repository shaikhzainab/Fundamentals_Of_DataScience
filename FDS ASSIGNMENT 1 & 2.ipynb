# Assignment-1

# 1)Create 4 list with name EmpID, FName, LName, Designation, Basic_Salary with minimum 5 records in each list. 

```python
EmpID=[1,2,3,4,5]
FName=['Zainab','Zain','Gulshan','Azhar','Umar']
LName=['Shaikh','Khan','Baloch','Khan','Pathan']
Designation=['Lab Assistant','Data Scientist','Accountant','Data Engineer','Manager']
Basic_Salary=[10000,60000,45000,30000,76000]
```

# 2) Create one table of above list with name “Empdata”

```python
Empdata=zip(EmpID,FName,LName,Designation,Basic_Salary)

```

# 3) Convert EmpData as dataframe.

```python
import pandas as pd
df=pd.DataFrame(data=Empdata,columns=['Emp_ID','F_Name','L_Name','Designation','Basic_Salary'])
print(df)
```

# 4) Convert EmpData to csv file and read contents of same csv file.

```python
df.to_csv('EmpData')
df
```

```python
df=pd.read_csv('EmpData')
df
```

```python
writer=pd.ExcelWriter('empdata.xlsx',engine='xlsxwriter')
df.to_excel(writer,sheet_name='sheet',index=False)
writer.save()
```

```python
df=pd.read_excel('empdata.xlsx')
df
```

# 6) Convert EmpData to sqlite, create database with name Employee.db and read contents of file.

```python
import sqlite3

con=sqlite3.connect('Employee.db')
cur=con.cursor()

df.to_sql('EmpData4',con)

```

```python
cur.execute('select *from EmpData4')
print(cur.fetchall())
```

# 7) Retrieve all the records of Employee.db whose salary in the range 20000-35000.

```python
cur.execute('select * from EmpData4  where Basic_Salary between 20000 and 35000')
print(cur.fetchall())
```
